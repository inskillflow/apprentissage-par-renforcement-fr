## 2. Moteur de quiz interactif

Tapez `a`, `b`, `c` ou `d` pour chaque question. Tapez `skip` pour passer.

```python
def poser_quiz(questions, titre):
    print("=" * 70)
    print(titre)
    print("=" * 70)
    score = 0
    details = []
    for item in questions:
        print("\n" + item["q"])
        for cle in ("a", "b", "c", "d"):
            print(f"   {cle}) {item['options'][cle]}")
        rep = input("Votre réponse (a/b/c/d) : ").strip().lower()
        correct = item["reponse"]
        if rep == correct:
            print("   -> Correct !")
            score += 1
            details.append((item["q"], True, correct))
        else:
            print(f"   -> Faux. Bonne réponse : {correct}) {item['options'][correct]}")
            details.append((item["q"], False, correct))
    print("\n" + "-" * 70)
    print(f"Score {titre} : {score} / {len(questions)}")
    print("-" * 70)
    return score, details

print("Fonction poser_quiz prête.")
```

## 3. Quiz 1 — Monte Carlo (20 questions)

```python
score_mc, details_mc = poser_quiz(QUIZ_MONTE_CARLO, "Quiz 1 - Monte Carlo")
```

## 4. Quiz 2 — TD-Learning (20 questions)

```python
score_td, details_td = poser_quiz(QUIZ_TD_LEARNING, "Quiz 2 - TD-Learning")
```

```python
total = score_mc + score_td
sur = len(QUIZ_MONTE_CARLO) + len(QUIZ_TD_LEARNING)
pourcentage = 100 * total / sur

print("=" * 70)
print("SCORE FINAL")
print("=" * 70)
print(f"Monte Carlo  : {score_mc} / {len(QUIZ_MONTE_CARLO)}")
print(f"TD-Learning  : {score_td} / {len(QUIZ_TD_LEARNING)}")
print(f"TOTAL        : {total} / {sur}  ({pourcentage:.1f} %)")
print("-" * 70)
if pourcentage >= 80:
    print("Excellent ! Concepts maîtrisés.")
elif pourcentage >= 60:
    print("Bien. Revoyez les questions ratées ci-dessus.")
else:
    print("À retravailler : relisez les sections 2 à 6 du chapitre.")
```
