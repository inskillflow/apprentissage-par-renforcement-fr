## 2. Questions de réflexion ouvertes (à rédiger soi-même)

**Q8 — Pourquoi est-il critique d'utiliser une baseline R̄_t dans la mise à jour du Gradient Bandit ?**

<details><summary>Corrigé</summary>

Sans baseline, on renforce H(A_t) à chaque récompense positive, même si l'action est sous la moyenne : convergence lente, sensibilité à l'échelle, variance énorme du gradient. Avec (R_t - R̄_t), seules les récompenses supérieures à la moyenne renforcent l'action ; l'apprentissage devient invariant à l'échelle et stable.

</details>

**Q9 — Parmi Q(a), N(a), c, k pour un agent UCB : lesquels sont des paramètres (appris) et lesquels des hyperparamètres (fixés) ?**

<details><summary>Corrigé</summary>

Q(a) et N(a) = **paramètres** (mis à jour par l'algorithme). c et k = **hyperparamètres** (c choisi avant le lancement ; k défini par la structure du problème). Règle : si l'algo le calcule -> paramètre ; si vous le choisissez avant -> hyperparamètre.

</details>

**Q14 — SLA < 100 ms, ~1 M utilisateurs/jour, reco de news personnalisée : quelle approche ?**

<details><summary>Corrigé</summary>

Bandit **contextuel** : LinUCB ou Linear Thompson Sampling (modèle linéaire, inférence rapide, apprentissage en ligne). Éviter Deep RL (latence/overkill), test A/B (gaspillage), bandit non-contextuel (pas de personnalisation). Stack : Vowpal Wabbit / Microsoft Personalizer, évaluation off-policy (IPS/Doubly Robust), audit d'équité, fallback top-N populaire.

</details>

**Q15 — Différence bandit stationnaire vs non-stationnaire, et comment adapter epsilon-greedy ?**

<details><summary>Corrigé</summary>

Stationnaire : q*(a) fixes -> moyenne empirique (alpha = 1/n). Non-stationnaire : q*(a) dérivent -> **alpha constant** (ex. 0.1) pour pondérer davantage les récompenses récentes. Développement : Q_{n+1} = (1-alpha)^n Q_0 + somme alpha (1-alpha)^{n-i} R_i -> poids exponentiellement décroissant avec l'âge.

</details>
